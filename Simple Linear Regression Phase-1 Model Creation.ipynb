```python
import pandas as pad #Panda is a libarary which deals with dataset.
```

```python
#Load the input file data into dataset variable by using read_csv() method
dataset=pad.read_csv("D:\AI\Hope AI\Week_3_ML_Regression\Simple Linear Regression\Salary_Data.csv")
```

```python
#Print the dataset
dataset
```

```python
#Retrieve the particular column values from dataset and assign to independent variable
Independent=dataset[["YearsExperience"]]
```

```python
#Print Independent variable
Independent
```

```python
#Retrieve the output column values from dataset and assign to dependent variable
dependent=dataset[["Salary"]]
```

```python
#Print dependent values
dependent
```

```python
#Split the train and test dataset values by using train_test_split function which is from sklearn.model_selection libarary

from sklearn.model_selection import train_test_split
X_trainset,X_testset,Y_trainset,Y_testset=train_test_split(Independent,dependent,test_size=0.30,random_state=0)
```

```python
X_testset
X_trainset
Y_trainset
```

```python
#Here user LinearRegression from sklearn.liner_model package. LinearRegression  is class and created an object as "regressor". By using object,we call fit method.
#fit method used for train& create the model of input and output training data set and it calculates weight and bias
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_trainset,Y_trainset)
```

```python
#Weight calculation
weight=regressor.coef_
weight
```

```python
#Bais Calculation
bais=regressor.intercept_
bais
```

```python
#Evaluate the model which means make sure our model whether good or bad.
y_pred=regressor.predict(X_testset)
```

```python
#r2score value should be  close 1 if model/brain is good. If r2score value close to 0 , model/brain is bad one which mean not trained properly.
from sklearn.metrics import r2_score
rScore=r2_score(Y_testset,y_pred)
rScore
```

```python
#Save the model using pickle package and model stored into separate file as bytes.
import pickle
filename="final_linear_model.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Load the saved model to test the input
loadedModel=pickle.load(open(filename,'rb'))
```

```python
#pass the experience into model and it returns predicted salary as output
result=loadedModel.predict([[10]])
```
